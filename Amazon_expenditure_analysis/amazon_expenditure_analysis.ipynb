```python
import pandas as pd
df=pd.read_csv("amazon-orders.csv")
df.head()
```

```python
df.shape
```

```python
#Data Cleaning
#Replace NAN values with zero and make chnages in orifinal data frame
df=df.fillna(0)
```

```python
df.head()
```

```python
#Data Cleaning
#Values of Total charged is string,So can't to mathmatical opertions. 
# Remove $ sign(Series.str.replace()) and convert string to float(astype(float))
df["Total Charged"]=df["Total Charged"].str.replace('$','').astype(float)
df.head()
```

```python
#Total amound of money spent on Amazon
df["Total Charged"].sum()
```

```python
#Average Spend
df["Total Charged"].mean()
```

```python
#Max Spend
df["Total Charged"].max()
```

```python
#Meadian Spend
df["Total Charged"].mean()
```

```python
#Tax Paid
df["Tax Charged"]=df["Tax Charged"].str.replace('$','').astype(float)
df.head()
```

```python
df["Tax Charged"].sum()
```

```python
df["Tax Charged"].sum() / df["Total Charged"].sum()
```

```python
#Analyzing Amazon Spending Over Time
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
%matplotlib inline
df.plot.bar(x='Order Date', y='Total Charged', rot=90,figsize=(30,30))
```

```python
#Above graph contains multiple entries for same day if there are multiple orders in 1 daya which is incorrect
daily_orders=df.groupby('Order Date').sum()["Total Charged"]
daily_orders.head()
```

```python
daily_orders.plot.bar(figsize=(20,10),color='green')
```

```python
import matplotlib.pyplot as plt

daily_orders=df.groupby('Order Date').agg({'Total Charged':'sum','Tax Charged':'sum'})
daily_orders.head()
```

```python

daily_orders.plot.bar(figsize=(20,10),color=['red','cyan'])
```
